# Visualização do Fuzz

```python
file_name = r'training\fuzz.csv'
fuzz_file = pd.read_csv(file_name)
normal_file = pd.read_csv(r'training\normal.csv')

ids_false = fuzz_file['ID'].unique()
print(ids_false, len(ids_false), "TOTAL")
```

```python
real_all = []
for id in ORDER:
	real_values = fuzz_file[(fuzz_file['ID'] == id) & (fuzz_file['NORMAL'] == True)]['DATA']
	if (not real_values.empty):
		real_values = np.unique(np.vectorize(lambda x: len(x))(real_values), return_counts= True)	
		real_all.append(real_values)
		print(f"{id}: {real_values[0]} Size {real_values[1]} Amount")
	
```

```python
false_all = []
for id in ids_false:
	false_values = fuzz_file[(fuzz_file['ID'] == id) & (fuzz_file['NORMAL'] == False)]['DATA']
	false_values = np.unique(np.vectorize(lambda x: len(x))(false_values), return_counts = True)
	false_all.append(false_values)
	print(f'{id}: Tamanho {false_values[0]} | Quantidade {false_values[1]}')	
```

```python
soma_percentage = 0
for false, id in zip(false_all, ids_false):
	print(f'{id}:', end='')
	total = 0
	length, value = false[0], false[1]
	if (id in ORDER):
		index = ORDER.index(id)
		for l, v in zip(length, value):
			if (l == real_all[index - 1][0]): continue
			total += v
	else: 
		for l, v in zip(length, value):
			total += v
	
	everything = (total)/(false[1].sum())
	soma_percentage += everything
	print(f' {everything * 100}%')
print("TOTAL ACERTOS: ", soma_percentage * 100 / len(ids_false), '%')
```
